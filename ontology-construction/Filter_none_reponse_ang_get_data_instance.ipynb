```python
# Extract column names from the 7th column to the last one in the second DataFrame
columns_of_interest = df.columns[6:].tolist()
```

```python
df1 = codebook[codebook['Question_Code'].isin(columns_of_interest)]

# Ensure the Answer_code column is numeric, coerce errors to NaN
df1['Answer_code'] = pd.to_numeric(df1['Answer_code'], errors='coerce')

# Drop rows where Answer_code is NaN (i.e., non-integer values or missing values)
df1 = df1.dropna(subset=['Answer_code'])

# Convert Answer_code to integer (since we know it is now numeric and not NaN)
df1['Answer_code'] = df1['Answer_code'].astype(int)

# Filter the DataFrame to include only rows where Answer_code is less than or equal to 20
filtered_df1 = df1[df1['Answer_code'] <= 20]

# List of words to filter out in the Answer_meaning column
irrelevant_words = ["no", "not", "other", "others", "don't"]

# Function to check if any of the irrelevant words are in the answer
def contains_irrelevant_words(answer):
    if isinstance(answer, str):
        return any(word in answer.lower() for word in irrelevant_words)
    return False

# Filter out rows with irrelevant words in Answer_meaning
filtered_df1 = filtered_df1[~filtered_df1['Answer_meaning'].apply(contains_irrelevant_words)]

# Save the filtered DataFrame to a new CSV file for manual review
filtered_df1.to_csv('filtered_answer_code_less_than_20_no_irrelevant_words.csv', index=False)

# Display a message indicating the file has been saved
print("Filtered data has been saved to 'filtered_answer_code_less_than_20_no_irrelevant_words.csv'")
```
